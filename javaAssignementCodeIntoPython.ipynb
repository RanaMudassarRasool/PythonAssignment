```python
from abc import ABC, abstractmethod

# No 1 - Abstract superclass
class Insurance(ABC):
    # 3.1 - Fields for superclass
    def __init__(self):
        self.policyNumber = ""
        self.policyHolderName = ""
        self.policyAmount = 0.0

    # 5.1 - Abstract methods
    @abstractmethod
    def calculatePremium(self):
        pass

    @abstractmethod
    def printPolicy(self):
        pass

# No 2 - Interface
class Transaction:
    @abstractmethod
    def deposit(self, amount):
        pass

    @abstractmethod
    def withdraw(self, amount):
        pass

# No 3 - Subclasses
# 3.2 Subclass
class HealthInsurance(Insurance):
    # 3.2 Fields for subclass
    def __init__(self):
        super().__init__()
        self.policyHolderAge = 0
        self.policyHolderGender = ""
        self.smoker = False

    # 5. User-defined methods
    def calculatePremium(self):
        self.policyAmount *= 0.03

    def printPolicy(self):
        print("Health Insurance Policy")

    def setPolicyHolderAge(self, policyHolderAge):
        self.policyHolderAge = policyHolderAge

    def setPolicyHolderGender(self, policyHolderGender):
        self.policyHolderGender = policyHolderGender

    def setSmoker(self, smoker):
        self.smoker = smoker

# 3.2 Subclass
class CarInsurance(Insurance):
    # 3.2 Fields for subclass
    def __init__(self):
        super().__init__()
        self.carModel = ""
        self.carMake = ""
        self.carYear = 0

    # 5. User-defined methods
    def calculatePremium(self):
        self.policyAmount *= 0.05

    def printPolicy(self):
        print("Car Insurance Policy")

    def setCarModel(self, carModel):
        self.carModel = carModel

    def setCarMake(self, carMake):
        self.carMake = carMake

    def setCarYear(self, carYear):
        self.carYear = carYear

# 3.2 Subclass
class HomeInsurance(Insurance):
    # 3.2 Fields for subclass
    def __init__(self):
        super().__init__()
        self.propertyType = ""
        self.propertyAddress = ""
        self.propertyValue = 0.0

    # 5. User-defined methods
    def calculatePremium(self):
        self.policyAmount *= 0.02

    def printPolicy(self):
        print("Home Insurance Policy")

    def setPropertyType(self, propertyType):
        self.propertyType = propertyType

    def setPropertyAddress(self, propertyAddress):
        self.propertyAddress = propertyAddress

    def setPropertyValue(self, propertyValue):
        self.propertyValue = propertyValue

# No 8 - Encapsulation
class BankAccount(Transaction):
    def __init__(self):
        self.balance = 0.0

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
        else:
            print("Insufficient funds.")

if __name__ == "__main__":
    # Create instances of each subclass
    healthPolicy = HealthInsurance()
    carPolicy = CarInsurance()
    homePolicy = HomeInsurance()
    # Set policy information
    healthPolicy.policyNumber = "HI1234"
    healthPolicy.policyHolderName = "John Doe"
    healthPolicy.policyAmount = 10000
    healthPolicy.setPolicyHolderAge(30)
    healthPolicy.setPolicyHolderGender("Male")
    healthPolicy.setSmoker(False)

    carPolicy.policyNumber = "CI5678"
    carPolicy.policyHolderName = "Jane Smith"
    carPolicy.policyAmount = 5000
    carPolicy.setCarMake("Toyota")
    carPolicy.setCarModel("Camry")
    carPolicy.setCarYear(2015)

    homePolicy.policyNumber = "HIJ910"
    homePolicy.policyHolderName = "Bob Johnson"
    homePolicy.policyAmount = 25000
    homePolicy.setPropertyType("Single Family Home")
    homePolicy.setPropertyAddress("123 Main St.")
    homePolicy.setPropertyValue(200000)

    # Calculate premiums and print policies
    healthPolicy.calculatePremium()
    healthPolicy.printPolicy()
    print("Policy Amount: " + str(healthPolicy.policyAmount))

    carPolicy.calculatePremium()
    carPolicy.printPolicy()
    print("Policy Amount: " + str(carPolicy.policyAmount))

    homePolicy.calculatePremium()
    homePolicy.printPolicy()
    print("Policy Amount: " + str(homePolicy.policyAmount))

```
